```python
# Import required depencencies: 
import pandas as pd
from sqlalchemy import create_engine
from sqlalchemy import inspect
import config
```

### Store CSV Files into DataFrame

```python
# Store first CSV file (Bitcoin) to data frame (Bitcoin_df):
Bitcoin_file = "Data/Bitcoin.csv"
Bitcoin_df = pd.read_csv(Bitcoin_file)
Bitcoin_df
```

```python
# Store first CSV file (Ethereum) to data frame (Ethereum_df):
Ethereum_file = "Data/Ethereum.csv"
Ethereum_df = pd.read_csv(Ethereum_file)
Ethereum_df
```

```python
# Store first CSV file (Tether) to data frame (Tether_df):
Tether_file = "Data/Tether.csv"
Tether_df = pd.read_csv(Tether_file)
Tether_df
```

```python
# Add a "Currency_Type" column with all fields equal to "Bitcoin"
Bitcoin_df["Currency_Type"] = "Bitcoin"
```

```python
# Add a "Currency_Type" column with all fields equal to "Ethereum"
Ethereum_df["Currency_Type"] = "Ethereum"
```

```python
# Add a "Currency_Type" column with all fields equal to "Tether"
Tether_df["Currency_Type"] = "Tether"
```

### Clean DataFrame

```python
# Drop NA entries:
Bitcoin_df.dropna(inplace=True)
Ethereum_df.dropna(inplace=True)
Tether_df.dropna(inplace=True)
```

```python
# Rename columns in data frame "Bitcoin_df" to match column names in data base:
Bitcoin_df = Bitcoin_df.rename(columns={'Vol.':'Volume', 'Change %':'Percentage_change'})
Bitcoin_df.reset_index(inplace=False)
Bitcoin_df
```

```python
# Rename columns in data frame "Ethereum_df" to match column names in data base:
Ethereum_df = Ethereum_df.rename(columns={'Vol.':'Volume', 'Change %':'Percentage_change'})
Ethereum_df
```

```python
# Rename columns in data frame "Tether_df" to match column names in data base:
Tether_df = Tether_df.rename(columns={'Vol.':'Volume', 'Change %':'Percentage_change'})
Tether_df
```

```python
# Find the min "Date" for all "Bitcoin" data frame:
Bitcoin_df['Date'].min()
```

```python
# Find the min "Date" for all "Ethereum" data frame:
Ethereum_df['Date'].min()
```

```python
# Find the min "Date" for all "Tether" data frame:
Tether_df['Date'].min()
```

```python
# Remove dates that are older than "2017-04-14" to stay aligned with the other data frames:
mask_bitcoin = (Bitcoin_df['Date'] >= "2017-04-14") & (Bitcoin_df['Date'] <= "2021-08-24")
Bitcoin_df = Bitcoin_df.loc[mask_bitcoin]
Bitcoin_df
```

```python
# Remove dates that are older than "2017-04-14" to stay aligned with the other data frames:
mask_ethereum = (Ethereum_df['Date'] >= "2017-04-14") & (Ethereum_df['Date'] <= "2021-08-24")
Ethereum_df = Ethereum_df.loc[mask_ethereum]
Ethereum_df
```

```python
# Concatenate all 3 data frames into one data frame called "final_df"
final_df = pd.concat([Bitcoin_df, Ethereum_df, Tether_df])
final_df
```

```python
# Reset index
final_df.reset_index(inplace=True)

# final_df.drop("SNo", axis=1, inplace=True)
final_df.drop("SNo", axis=1, inplace=True)

final_df.reset_index(drop=True)
```

```python
# Save final_df as "final_data.csv" under the Data folder
final_df.to_csv("Data/final_data.csv", index=False)
```
